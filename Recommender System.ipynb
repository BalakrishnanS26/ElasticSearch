```python
import requests

def get_es_indices():
    r = requests.get("http://elasticsearch:9200/_cat/indices?format=json")
    if r.status_code != 200:
        print("Error listing indices")
        return None
    else:
        indices_full = r.json()  # contains full metadata as a dict
        indices = []  # let's extract the names separately
        for i in indices_full:
            indices.append(i['index'])
        return indices, indices_full
        
indices, indices_full = get_es_indices()
print(indices)
```

#### Defining the index and checking the current indices.

```python
def create_es_index(index, index_config):
    r = requests.put("http://elasticsearch:9200/{}".format(index),
                     json=index_config)
    if r.status_code != 200:
        print("Error creating index")
    else:
        print("Index created")
        
def delete_es_index(index):
    r = requests.delete("http://elasticsearch:9200/{}".format(index))
    if r.status_code != 200:
        print("Error deleting index")
    else:
        print("Index deleted")
```

#### Defining create index and defining delete index.

```python
def fling_message(index, doctype, msg):
    r = requests.post("http://elasticsearch:9200/{}/{}".format(index, doctype),
                      json=msg)
    if r.status_code != 201:
        print("Error sending message")
    else:
        print("message sent")
```

#### Following the lecture and using the define fling message.

```python
def change_format(dict):
    dict_format=dict
    dict_format['InvoiceNo']=str(dict['InvoiceNo'])
    dict_format['CustomerID']=str(dict['CustomerID'])
    dict_format['StockCodes']=str(dict['StockCodes'])
    dict_format['Descriptions']=dict['Descriptions']
    del dict_format['Quantities']
    del dict_format['StockCodes']
    del dict_format['UnitPrices']
    dict_format['InvoiceDate']=str(dict['InvoiceDate'])
    del dict_format['Country']
    del dict_format['InvoiceDate']
    return dict_format
```

#### Converting the dataset you created to a dictionary format for further analysis.

```python
# delete if exists
indices, indices_full = get_es_indices()
if 'recommender_system' in indices:
    delete_es_index('recommender_system')
    

index_config = {
    "mappings": {
        "recommender": {  # document TYPE
            "properties": {
                "InvoiceNo": {"type": "string","index":"not_analyzed"},
                "CustomerID": {"type": "string","index":"not_analyzed"},
                "InvoiceDate": {"type": "date","index":"not_analyzed"},
                "StockCodes": {"type": "string","index":"not_analyzed"},
                "Descriptions": {"type": "string","index":"not_analyzed"}
            }
        }
    }
}

create_es_index('recommender_system', index_config)
```

#### Followed the lecture code to define the recommender system and specified the fields in the index. We are creating the index thereafter. 

```python
import pandas as pd
import requests

# read in CSV
df = pd.read_csv('Online_Retail.csv')

# group by invoice num
invoice_groups = df.groupby('InvoiceNo')

# iterate over each group (each invoice)
for invoice_name, invoice in invoice_groups:
    basket = {}
    stockcodes = []
    descriptions = []
    quantities = []
    unitprices = []
    # iterate over rows in this invoice dataframe
    for row_index, row in invoice.iterrows():
        # these fields are the same for each row, so doesn't matter if we keep overwriting
        basket['InvoiceNo'] = row['InvoiceNo']
        basket['CustomerID'] = row['CustomerID']
        basket['InvoiceDate'] = row['InvoiceDate']
        basket['Country'] = row['Country']
        # these fields are different for each row, so we append to lists
        stockcodes.append(row['StockCode'])
        descriptions.append(row['Description'])
        quantities.append(row['Quantity'])
        unitprices.append(row['UnitPrice'])
    basket['StockCodes'] = stockcodes
    basket['Descriptions'] = descriptions
    basket['Quantities'] = quantities
    basket['UnitPrices'] = unitprices
    basket=change_format(basket)
    #r=requests.post("")
    fling_message('recommender_system','recommender',basket)
 # now you should fling to elasticsearch!
    #print(basket)
```

#### Followed your code to iterate over each group (for each invoice), after which i apply the function (change_format) to change the basket to dictionary format. Once that is done, we start to send the message to the kibana using the function fling_message.

```python
r = requests.post("http://elasticsearch:9200/recommender_system/recommender",
                  json=basket)
if r.status_code != 201:
    print("Error sending message")
else:
    print("message sent")
```
